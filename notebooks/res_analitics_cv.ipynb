```python
X = pd.read_pickle('../results/hist_cv_3.pkl')
x = x.loc[x.sample_size == 30000]
fig = plt.figure(figsize=(15,40))
for i,sample in enumerate([1000,2000,5000,8000,10000,15000,20000,30000]):
    x = pd.read_pickle(f'../results/best_nn/hist_sample_{sample}.pkl')
    
    plt.subplot(len(X.sample_size.unique()),2,i*2+1)
    plt.title(f'IDGT cnn: sample {sample} - loss per epoch')
    plt.plot(x['loss'].values,label = 'train loss')
    plt.plot(x['val_loss'].values,label = 'validation loss')
    plt.legend(loc = 2)
    plt.subplot(len(X.sample_size.unique()),2,i*2+2)
    plt.title(f'IDGT cnn: sample {sample} - accuracy per epoch')
    plt.plot(x['accuracy'].values,label = 'train acc')
    plt.plot(x['val_accuracy'].values,label = 'validation acc')
    plt.legend(loc = 2)
```

```python
X = pd.read_pickle('../results/best_his_res_1.1.pkl')
x = x.loc[x.sample_size == 30000]
fig = plt.figure(figsize=(15,40))
for i,sample in enumerate(X.sample_size.unique()):
    x = X.loc[X.sample_size == sample]
    
    plt.subplot(len(X.sample_size.unique()),2,i*2+1)
    plt.title(f'nn : sample {sample} - loss per epoch')
    plt.plot(x['loss'].values[0],label = 'train loss')
    plt.plot(x['val_loss'].values[0],label = 'validation loss')
    plt.legend(loc = 2)
    plt.subplot(len(X.sample_size.unique()),2,i*2+2)
    plt.title(f'nn: sample {sample} - accuracy per epoch')
    plt.plot(x['accuracy'].values[0],label = 'train acc')
    plt.plot(x['val_accuracy'].values[0],label = 'validation acc')
    plt.legend(loc = 2)
```
